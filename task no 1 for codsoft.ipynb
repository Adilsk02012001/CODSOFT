```python
# TASK NO 01 FOR CODSOFT INTERNSHIP

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
dataset=pd.read_csv("dataset for codsoft task 1.csv")
dataset[dataset.Age.isnull()]
```

```python
male_survived_age_median = dataset.loc[(dataset['Sex'] == 'male') & (dataset['Survived'] == 1), 'Age'].median()
male_death_age_median = dataset.loc[(dataset['Sex'] == 'male') & (dataset['Survived'] == 0), 'Age'].median()
female_survived_age_median = dataset.loc[(dataset['Sex'] == 'female') & (dataset['Survived'] == 1), 'Age'].median()
female_death_age_median = dataset.loc[(dataset['Sex'] == 'female') & (dataset['Survived'] == 0), 'Age'].median()
```

```python
dataset.loc[(dataset['Sex'] == 'male') & (dataset['Survived'] == 1), 
'Age'] = dataset.loc[(dataset['Sex'] == 'male') & 
(dataset['Survived'] == 1), 'Age'].fillna(male_survived_age_median)
```

```python
dataset.loc[(dataset['Sex'] == 'male') & (dataset['Survived'] == 0),
'Age'] = dataset.loc[(dataset['Sex'] == 'male') & 
(dataset['Survived'] == 0), 'Age'].fillna(male_death_age_median)
```

```python
dataset.loc[(dataset['Sex'] == 'female') & (dataset['Survived'] == 1),
'Age'] = dataset.loc[(dataset['Sex'] == 'female') & (dataset['Survived'] == 1),
'Age'].fillna(female_survived_age_median)
```

```python
 dataset.loc[( dataset['Sex'] == 'female') & ( dataset['Survived'] == 0), 'Age'] = dataset.loc[( dataset['Sex'] == 'female') & 
( dataset['Survived'] == 0), 'Age'].fillna(female_death_age_median)
```

```python
age_groups = pd.cut(dataset['Age'], bins=range(0, 100, 10), include_lowest=True)
pd.crosstab(age_groups,dataset.Survived).plot(kind='bar')
plt.ylabel('No.of Passengers')
plt.title('Age over Survival')
plt.grid(color="blue", linestyle=":", alpha=0.5)
```

```python
pd.crosstab(dataset.Sex,dataset.Survived).plot(kind='bar')
plt.ylabel('No.of Passengers')
plt.title('Sex over Survival')
plt.grid(color="blue", linestyle=":", alpha=0.5)
```

```python
fare_groups = pd.cut(dataset['Age'], bins=range(0, 550, 10), include_lowest=True)
pd.crosstab(fare_groups,dataset.Survived).plot(kind='bar')
plt.ylabel('No.of Passengers')
plt.title('Fare over Survival')
plt.grid(color="black", linestyle=":", alpha=0.5)
```
